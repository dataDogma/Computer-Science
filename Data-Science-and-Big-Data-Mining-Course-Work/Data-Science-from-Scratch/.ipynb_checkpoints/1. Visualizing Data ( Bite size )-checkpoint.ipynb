### Matplotlib
-- --

Matplotlib is a built-in library which comes packaged with Python distributions. As a general note, it's much suited for producing visualization on local machine, or for self-research. For interactive visualization oriented for web, try D3.js.

**Using Matplotlib:**

I am using matplotlib with the _pyplot_ module. In the simplest sense, _pyplot maintains an internal state_, i.e.

+ We build visualizations step by step.


+ Once done, we can save it using `savefig()` or display using `show()` method.


**First Bite:**

Let's take it for a spin!

```python
"""
####################################################################
# Title: Making simple plots                                       #
# -----------------------------------------------------------------#
# About: A fictional GDP(gross domistic product)                   #
# X-axis: Time(years)                                              #
# Y-axis: GDP(in Billions of $)                                    #
####################################################################
"""

# Import the matplolib package with the pyplot module.
# Use "plt" as an "alias" to work easily when calling the methods.

from matplotlib import pyplot as plt

# Next, let's create some fictional data for our use as "python lists"

year = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.6, 1075.9, 2862.5, 5879.6, 10289.7, 14958.3]

# Create a line chart, years on X-axis, gdp on Y-axis
# Using our alias, we call the "plot" method.
# plot(), takes many arguments, but for now let's just pass in the two list's

plt.plot( year, gdp)

# Let's add some title and labels

plt.title("Nominal GDP")                      # Add a title to our plot
plt.xlabel("Time in years")                   # Add label to X-axis
plt.ylabel("Gdp in Billions of $")            # Add label to Y-axis

# Keeping in mind that "pyplot" method keeps an internal state, let's display our plot
# Notice, each step we've been so far, we have respected the flow of visualization.

plt.show()


```

**Second Bite**
-- --

In the most simplest sense, a data visualization workflow will consist of the following steps:

1. ETL(Extract, transform and load) the data.
2. Plot the data.
3. Add some more cosmetics to our visualization.
4. Display or save the visualization.

-- --

A good practice to make good visualizations stand out, proceed with step 3 by excluding it at first i.e. proceed with step 1, 2 and 4 at first and then go with step 3.

Let's now add some cosmetics to our visualization!

```python
# Add more arguments to "plot" method, some of them are:
# "color" - pretty self-explainatory!
# "marker" - color filled blobs(markers) which represent distinct point of data in question.
# "linestyle" - pretty self-explainatory!

plt.plot(year, gdp, color='green', marker='o', linestyle='solid')

# let's display an improved plot
plt.show()
```

### Bar Chart
-- --

**Introduction:**

A bar chart is a good choice when we wan't to show how multiple quantities varies among some _discrete_ set of items.

**Example** 

A plot of _how many Academy Awards were won?_ by a set of _discrete variety of movies_.

```python
"""
######################################################
# Title: My favorite Movies                          #
# ---------------------------------------------------#
# About: How many Academy Awards won by my favourite #
#        movies.                                     #
# Labels: X-axis - Favorite Movies.                  #
#         Y-axis - No. of Academy Awards won.        #
######################################################
"""

# Step 1: Create some data

movies = [ "Forest Gump", "TSOL", "Gladiator", "Pianist", "Dark Kngt", "Hurt-locker"]
num_oscars = [ 6, 5, 5, 3, 2, 6]

# Width of bars : 0.8 by default, we change it to 0.1 to the left coordinates
# So that each bar is nicely centered

width_xaxis = [ i + 0.1 for i, _ in enumerate( movies )]

# Plot the bars using "bar" method,
# X-axis is "width_xaxis" and Y-axis[heights] is "num_oscars".

plt.bar( width_xaxis, num_oscars, color = 'teal')


# Add Title and labels

plt.title("My Favourite Movies")
plt.ylabel("# of Academy Awards")

# Plot each bar with an appropriate movie label using xticks() method
plt.xticks([i + 0.5 for i, _ in enumerate(movies)], movies)

# Display the plot
plt.show()
```
